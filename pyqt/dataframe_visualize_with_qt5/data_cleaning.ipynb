```python
special_ch = ".,:;'-_/"
column_list = ['Product', 'Issue', 'Company']
for x in special_ch:
  for y in range(21349):
    for k in column_list:
      if cropped_df[k][y].find(x) != -1:
        cropped_df[k][y] = cropped_df[k][y].replace(x,"")
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
column_list = ['Product', 'Issue', 'Company']
for k in column_list:
  cropped_df[k] = cropped_df[k].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```
